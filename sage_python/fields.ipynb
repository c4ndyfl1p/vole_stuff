## Polynomial ring over $F_{p}[X]$

1. Now we create polynomials $f$ whose coefficients lie in the field $F_p$.



$f(t)=3 t^2+5t+11 \in F_{17}[X] $

$f(t)=12 t^9 -3 t^2+5t+11 \in F_{17}[X] $


Here:

$12, 3,5,11∈F_{17}$ 




Question: what is the difference between a polynomial field over F_p and polynomial over F_p

A polynomial with coeefiencits in $F_p$ is a polynomial.

2. The set of all such polynomials is the polynomial ring. This can be infinite as degree of the indeterminent $t$ can be anything.


```python
R, t = PolynomialRing(Fp, 't').objgen()

f = Fp(2)*(t**2) + 5*t + 1
g = Fp(2)*(t**2) + 13*t + 1

sum_poly = f+g
prod_poly = f*g #oh degree grows.

print(f)
print(g)
print(f"Sum= {f+g}")
print(f"Product= {prod_poly}, Degree= {prod_poly.degree()}" ) #note: the degree grows

print(g.parent())

#list all elemtns of the ring?
# cdannot because this is an infintie ring
```

```python
# test ring axioms
assert f + g == g + f # commutativity of addition
assert f * g == g * f # commutativity of multiplication
assert (f + g) + f == f + (g + f) # associativity
assert (f * g) * f == f * (g * f) # associativity
assert f * (g + f) == f * g + f * f # distributivity
#also test the additive and multiplicative identities
assert f + 0 == f # additive identity
assert f * 1 == f # multiplicative identity
# test additive inverses


# multiplicative inverses (for non-zero elements)
if f != 0:
    f_inv = 1/f
    print(f"f: {f}, f_inv: {f_inv}")
    assert f * f_inv == 1 # multiplicative inverse

```

### Quotient Ring of $Q[x] = F_p[X]/f(x)$

We can also make the ring finite by reducing everyuthing a modulo polynomial. This reduction polynomial deos nopt ahve to be irreducible. If it is, we get an extention field(i think)

```python
R, x = PolynomialRing(Fp, 'x').objgen()
irr_poly = x**3 + x + 1
print(irr_poly.is_irreducible())

irr_poly = x**2 + x + 1
print(irr_poly.is_irreducible())

Q = QuotientRing(R, irr_poly, 'a')

print(list(Q))
print(Q.parent())
```

### Extention field $GF_{p^r} = F_{17}[X]/(f(x))$

We can now take the above ring, and make it finite, and also a field, by reducing each element by an irreducible polynomial $f(x)$. This is a special kind of a quotient ring.

For example: $GF(2^8)=F_2​[x]/(f(x)=x^8 + x^4 + x^3 + x + 1)$

This looks like a polynomial with co-efficients in the base field F_p (note:coeffs are _not_ in $F_p^r$), but $r$ number of such coeefients.

$f = a_0 + a_1(X) + ... + a(r-1)(X^{r-1})$

The size of the field thus is $ \underbrace{p \cdot p \cdot ... \cdot p}_{r} = p^r$





```python
Fpr = GF(2**2, 't') #t can be used to create elements in the extension field
# # print the irriducible polynomial used to construct the extension field
print(f"moulus of field is {Fpr.modulus()}")

# inittialise the inteterminate element in Fp^2
d = Fpr.gen()
print(f"generator of the field is {d}") # now we can use d as well to create eleemtns in the extension field

f = Fpr.random_element()
g = Fpr.random_element()

sum_poly = f+g
prod_poly = f*g #degree does not grow in extension fields
print(f"f: {f} \ng: {g} \nf+g: {sum_poly} \nf*g: {prod_poly}")
print(prod_poly.parent())
# print(type(prod_poly))

print(list(Fpr))
```
